One key drawback of decision trees is their susceptibility to variance. The resulting tree relies heavily on the distribution of data within the training data, such that different subsets from the same training data can create vastly different trees. One key component that gives rise to this variance is the nature of entropy - the measure of disorder within a set of examples. As seen by the equation for entropy, the value depends entirely on the probability of each feature value. Evidently, data with high variance can lead to vastly different entropies between different subsets.

 
We hypothesized that average entropy between multiple subsets would provide a more reliable metric of discriminative features that are resilient to variance. To test our hypothesis, we examined the mean entropy between multiple subsets of training data during our calculation of information gain/gain ratio/entropy. 
 
Methodology:
Using m as a hyperparameter, we divided the training data of n examples into m disjoint stratified subsets. For each subset, we randomly added (n-m)/2 examples from the other subsets while maintaining the class distribution. Ultimately, each subset consisted of n/m + (n-m)/2 examples. In this way, we ensured that every example was used at least once, while creating some randomness by randomly selecting additional examples. We preserve the class distribution to ensure both positive and negatives are weighted proportionally. 

We used each subset to calculate the mean gain ratio. The mean gain ratio was used to represent each split criterion, and the criterion resulting in the highest mean gain ratio was selected as the next node on the decision tree. By injecting greater variability into the training set, we aimed to identify features which consistently performed well.

An important consideration is the value of m selected. We experimented with multiple values of m by grid search to gain insight into how subset size and selection affects resilience to variance. Intuitively, we expected m values that are too small to have high variance and poor generalizability because the training data would not be too sparse. In contast, large m values would fall subject to the opposite issue of being too similar to the full training set. This would also lead to poor generalizability becauses the variance would not be adequately accounted for. 

A graph of performance with varying values of m on the easy voting dataset with maximum depth 2, and the difficult spam dataset with a maximum depth of 4 is shown below. The performance of our standard decision tree on these two datasets is shown as well



```python
import mldata
import dtree
import random
import pandas
def get_random_subset(example_set: mldata.ExampleSet, prop: float):
    random.shuffle(example_set)
    return mldata.ExampleSet([example_set[e] for e in range(int(len(example_set) * prop))])

voting = mldata.parse_c45('voting')
experimental_results = dtree.main2(voting, 0, 2, -1) #-1 in the last position runs our experimental metric 
voting_res = dtree.main2(voting, 0, 2, 0) # runs information gain


df=pd.DataFrame({'partitions': [3, 5, 7, 10], 'experimental': experimental_results, 'voting' : [voting_res, voting_res, voting_res, voting_res]})
plt.plot( 'partitions', 'experimental', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'partitions', 'voting', data=df, marker='|', color='orange', linewidth=2)
plt.legends()
```

In voting, our selection method performs slightly better than the traditional method with most m values, though both select Repealing the Job Killing Care Law Act as the first node. It is likely the second node is different, leading to the superior performance

```python

spam = mldata.parse_c45('spam')

spam_small = get_random_subset(spam, 0.01)
experimental_results = dtree.main2(spam_small, 0, 2, -1) #-1 in the last position runs our experimental metric 
spam_results = dtree.main2(spam_small, 0, 2, 0) # runs information gain

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.DataFrame({'partitions': [3, 5, 7, 10], 'experimental': experimental_results, 'spam': [spam_results, spam_results, spam_results, spam_results]})
plt.plot( 'partitions', 'experimental', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'partitions', 'spam', data=df, marker='x', color='olive', linewidth=2)
plt.legend()
```

In spam, we see comparable performance. Notably, the run time with this was extremely slow, because we must calculate information gain m times for every partition, for every test. Consequently, we randomly use a subset of examples from the spam set to obtain realistic runtime (still took hours). Thus, we see the improved accuracy comes at the cost of significantly longer runtime. Whether this increase in accuracy is worth the computational time depends on the application, but the results do indicate our method is well suited for addrerssing an issue with the variability of decision trees. Both tests select geodistance, so may be that geodistance is simply a distinctly superior feature. The maximum average accuracy with our experimental metric is 0.0.6144519015659955 compared to 0.6144608501118568 for the traditional decision tree. Thus, for complicated datasets, the results seem comparable between our stochastic information gain and traditional information gain.
